```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('C:\\Users\\user\\Desktop\\fpa\\train.csv', index_col='Row ID')
data
```

```python
data[data['Postal Code'].isnull()]
```

```python
data['Postal Code'] = data['Postal Code'].fillna('05049')
data.info()
```

```python
data['Order Date'] = pd.to_datetime(data['Order Date'], format='%d/%m/%Y')
data['Ship Date'] = pd.to_datetime(data['Ship Date'], format='%d/%m/%Y')
```

```python
data['Order Month'] = data['Order Date'].dt.to_period('M')
data['Delivery Time'] = (data['Ship Date'] - data['Order Date']).dt.days
data
```

```python
sale_sum = data['Sales'].sum()
sale_mean = data['Sales'].mean()
unique_customers = data['Customer ID'].nunique()
unique_product = data['Product ID'].nunique()
print(sale_sum.round(2), sale_mean.round(2), unique_customers, unique_product, sep=', ')
```

```python
salesbycategory = data.groupby(['Category', 'Sub-Category'])['Sales'].sum().sort_values(ascending=False)
sales_region = data.groupby('Region')['Sales'].sum().sort_values(ascending=False)
sales_city = data.groupby('City')['Sales'].sum().sort_values(ascending=False)
sales_state = data.groupby('State')['Sales'].sum().sort_values(ascending=False)
sales_segment = data.groupby('Segment')['Sales'].sum().sort_values(ascending=False)
salesbycategory
```

```python
sales_region
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (10, 6)
```

```python
# 1. Продажи по категориям и подкатегориям — топ-10
category_plot_data = salesbycategory.head(10).reset_index()

plt.figure(figsize=(10, 6))
sns.barplot(data=category_plot_data, y='Sub-Category', x='Sales', hue='Category', dodge=False)
plt.title('🔝 Топ-10 подкатегорий по продажам')
plt.xlabel('Продажи')
plt.ylabel('Подкатегория')
plt.tight_layout()
plt.show()
```

```python
region_plot_data = sales_region.reset_index()

plt.figure(figsize=(8, 5))
sns.barplot(data=region_plot_data, x='Region', y='Sales', palette='Set2')
plt.title('📍 Продажи по регионам')
plt.xlabel('Регион')
plt.ylabel('Продажи')
plt.tight_layout()
plt.show()
```

```python
segment_plot_data = sales_segment.reset_index()

plt.figure(figsize=(8, 5))
sns.barplot(data=segment_plot_data, x='Segment', y='Sales', palette='Set1')
plt.title('👥 Продажи по сегментам клиентов')
plt.xlabel('Сегмент')
plt.ylabel('Продажи')
plt.tight_layout()
plt.show()
```

```python
data['YearMonth'] = data['Order Date'].dt.to_period('M')
data['Month'] = data['Order Date'].dt.month
```

```python
monthly_sales = data.groupby('YearMonth')['Sales'].sum().reset_index()
monthly_sales['YearMonth'] = monthly_sales['YearMonth'].astype(str)
```

```python
seasonility = data.groupby('Month')['Sales'].mean().reset_index()
seasonility
```

```python
shipping_delay = data.groupby('YearMonth')['Delivery Time'].mean().reset_index()
shipping_delay['YearMonth'] = shipping_delay['YearMonth'].astype(str)
```

```python
monthly_sales.head(), seasonility.head(), shipping_delay.head()
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=monthly_sales, x='YearMonth', y='Sales', marker='o')
plt.xticks(rotation=45)
plt.title('📅 Продажи по месяцам')
plt.xlabel('Месяц')
plt.ylabel('Продажи')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.lineplot(data=seasonility, x='Month', y='Sales', marker='o', color='orange')
plt.title('📈 Сезонность: средние продажи по месяцам')
plt.xlabel('Месяц (1–12)')
plt.ylabel('Средние продажи')
plt.xticks(range(1, 13))
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=shipping_delay, x='YearMonth', y='Delivery Time', marker='o', color='green')
plt.xticks(rotation=45)
plt.title('🚚 Среднее время доставки по месяцам')
plt.xlabel('Месяц')
plt.ylabel('Средняя задержка (дней)')
plt.tight_layout()
plt.show()
```
